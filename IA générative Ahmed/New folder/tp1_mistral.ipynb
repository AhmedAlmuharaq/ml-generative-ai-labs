```python
# TP1 - Step 3: First invocation of a Mistral model with LangChain
# Goal: Send one simple question to the model and print the answer.
# What you learn: How to load API keys securely and call llm.invoke().

import os
from dotenv import load_dotenv
from langchain_mistralai.chat_models import ChatMistralAI
from langchain_core.messages import HumanMessage

# Load environment variables from .env into the current Python process
load_dotenv(override=True)

# Read the API key from environment variables (do NOT hardcode it in code)
api_key = os.getenv("MISTRAL_API_KEY")
if not api_key:
    raise ValueError("Missing MISTRAL_API_KEY. Put it in .env or set it as an env var.")

# Create the Mistral LLM client
# temperature=0 => more deterministic answers (good for factual questions)
llm = ChatMistralAI(
    model="mistral-large-latest",
    temperature=0,
    api_key=api_key
)

# Define the user question (HumanMessage = user role)
question = "who are you?"

# Invoke the model (LangChain expects a list of messages for chat models)
response = llm.invoke([HumanMessage(content=question)])

# Print the model output
print("Question:", question)
print("Answer:", response.content)

# (Optional) Print token usage metadata if available (useful for Ex 3.1 cost estimation)
print("usage_metadata:", getattr(response, "usage_metadata", None))
```

```python
# TP1 - Exercise 3.1: Inspect token usage and estimate cost
# Goal: Print response.usage_metadata and compute approximate cost using public pricing.
# What you learn: How to read input/output tokens and translate them into money.

import os
from dotenv import load_dotenv
from langchain_mistralai.chat_models import ChatMistralAI
from langchain_core.messages import HumanMessage

load_dotenv(override=True)
api_key = os.getenv("MISTRAL_API_KEY")
if not api_key:
    raise ValueError("Missing MISTRAL_API_KEY.")

llm = ChatMistralAI(model="mistral-large-latest", temperature=0, api_key=api_key)

question = "Explain TCP vs UDP in simple terms."
response = llm.invoke([HumanMessage(content=question)])

print(response.content)

usage = getattr(response, "usage_metadata", None)
print("usage_metadata:", usage)

# IMPORTANT: Replace these with the real public prices for your model
# Example format: price per 1K tokens (not real numbers)
PRICE_INPUT_PER_1K = 0.0
PRICE_OUTPUT_PER_1K = 0.0

def estimate_cost(input_tokens: int, output_tokens: int) -> float:
    # Cost = (input_tokens/1000)*input_price + (output_tokens/1000)*output_price
    return (input_tokens / 1000.0) * PRICE_INPUT_PER_1K + (output_tokens / 1000.0) * PRICE_OUTPUT_PER_1K

if usage:
    input_tokens = usage.get("input_tokens", 0)
    output_tokens = usage.get("output_tokens", 0)
    cost = estimate_cost(input_tokens, output_tokens)

    print(f"input_tokens: {input_tokens}")
    print(f"output_tokens: {output_tokens}")
    print(f"estimated_cost: {cost:.6f} (currency depends on your pricing)")
else:
    print("No usage_metadata returned by the provider in this response.")
```

```python
# TP1 - Step 4: Chains and structured prompts
# Goal: Build a chain: prompt template -> model -> output parser.
# What you learn: How LangChain composes steps with the | operator.

import os
from dotenv import load_dotenv
from langchain_mistralai.chat_models import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv(override=True)
api_key = os.getenv("MISTRAL_API_KEY")
if not api_key:
    raise ValueError("Missing MISTRAL_API_KEY.")

model = ChatMistralAI(model="mistral-large-latest", temperature=0, api_key=api_key)

# Prompt with a variable {topic}
prompt = ChatPromptTemplate.from_template("Tell me a joke about: {topic}")

# Output parser converts model output into a simple string
parser = StrOutputParser()

# Create the chain (prompt -> model -> parser)
chain = prompt | model | parser

topics = ["firefighters", "police"]

for topic in topics:
    print("Topic:", topic)
    print(chain.invoke({"topic": topic}))
    print("-" * 20)
```

```python
# TP1 - Exercise 4.1: Dynamic list of topics + empty-string protection
# Goal: Let the user type topics and prevent sending empty input to the model.
# What you learn: Basic input validation + dynamic invocation of the chain.

import os
from dotenv import load_dotenv
from langchain_mistralai.chat_models import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv(override=True)
api_key = os.getenv("MISTRAL_API_KEY")
if not api_key:
    raise ValueError("Missing MISTRAL_API_KEY.")

model = ChatMistralAI(model="mistral-large-latest", temperature=0, api_key=api_key)
prompt = ChatPromptTemplate.from_template("Tell me a joke about: {topic}")
parser = StrOutputParser()
chain = prompt | model | parser

# User enters comma-separated topics
raw = input("Enter topics (comma-separated): ").strip()
topics = [t.strip() for t in raw.split(",") if t.strip()]

# Block empty list
if not topics:
    raise ValueError("No valid topics provided (empty input).")

for topic in topics:
    print(chain.invoke({"topic": topic}))
    print("-" * 20)
```

```python
# TP1 - Exercise 4.3: Compare temperatures on the same topics
# Goal: Build multiple models with different temperatures and compare outputs.
# What you learn: Temperature is a model parameter, so easiest is to instantiate different models.

import os
from dotenv import load_dotenv
from langchain_mistralai.chat_models import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv(override=True)
api_key = os.getenv("MISTRAL_API_KEY")
if not api_key:
    raise ValueError("Missing MISTRAL_API_KEY.")

prompt = ChatPromptTemplate.from_template("Tell me a joke about: {topic}")
parser = StrOutputParser()

def run_with_temperature(topic: str, temperature: float) -> str:
    model = ChatMistralAI(model="mistral-large-latest", temperature=temperature, api_key=api_key)
    chain = prompt | model | parser
    return chain.invoke({"topic": topic})

topics = ["students", "pip", "debugging"]

for t in topics:
    print(f"=== Topic: {t} | temp=0 ===")
    print(run_with_temperature(t, 0.0))
    print(f"\n=== Topic: {t} | temp=0.7 ===")
    print(run_with_temperature(t, 0.7))
    print("-" * 40)
```

```python
# TP1 - Step 5: Hierarchical prompt (system + user) + usage metadata
# Goal: Use a system message to control style/role and inspect usage_metadata.
# What you learn: How system instructions influence output style and token usage.

import os
from dotenv import load_dotenv
from langchain_mistralai.chat_models import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate

load_dotenv(override=True)
api_key = os.getenv("MISTRAL_API_KEY")
if not api_key:
    raise ValueError("Missing MISTRAL_API_KEY.")

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a world-class technical documentation writer. Be concise."),
    ("user", "{input}")
])

llm = ChatMistralAI(model="mistral-large-latest", api_key=api_key)
chain = prompt | llm

result = chain.invoke({"input": "What is mistral-large-latest?"})

print(result.content)
print(getattr(result, "usage_metadata", None))
```
